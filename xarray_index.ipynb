# Index and selecting data

```python
import xarray as xr
import numpy as np
import pandas as pd
```

```python
da = xr.DataArray(
    np.random.rand(4, 3),
    [
        ("time", pd.date_range("2000-01-01", periods=4)),
        ("space", ["IA", "IL", "IN"]),
    ],
)
da
```

```python
da[:2] # 左闭右开的直接引用
```

```python
da.coords
```
